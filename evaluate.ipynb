```python
import pandas as pd
```

```python
data = pd.read_csv("out/summary.csv")
```

```python
data["solver_duration"] = data["solver_end"] - data["solver_start"]
```

```python
data
```

```python
groups = data.groupby(["id"])
```

```python
REFERENCE = "fwd"

extended_groups = []

for name, gr in groups:
    reference_duration = gr[gr["conf"] == REFERENCE]["solver_duration"].mean()
    gr["relative_solver_duration"] = gr["solver_duration"] / reference_duration
    extended_groups.append(gr)
    
extended_data = pd.concat(extended_groups)
extended_data
```
